```python
import pandas as pd
```

```python
df = pd.read_csv("IIT_Training_Data")
```

```python
df = df.drop(columns=["Unnamed: 0"])
```

```python
df
```

```python
df.dtypes
```

```python
def convert(value):
    if type(value) == str:
        value = value.strip('P')
        value= float(value)
    return float(value)

```

```python
for i in range(0,len(df["Opening"])):
    
    df["Opening"][i] = convert(df["Opening"][i])
    
     
    
for i in range(0,len(df["Closing"])):
    
    df["Closing"][i] = convert(df["Closing"][i])
    
     
    
    
```

```python
df["Closing"] = df["Closing"].astype(int)
df["Opening"] = df["Opening"].astype(int)
```

```python
df.to_csv("training")
```
